```python
def add_json(filename):
    import json
    
    #json파일 읽기
    with open('words.json','r',encoding='utf-8') as f:
        json_data = json.load(f)
    
    #원래 있던 json 파일
    #print(json.dumps(json_data,ensure_ascii=False,indent='\t'))
    
    #매장 list만들기
    shop_list=list()
    with open(filename,'r',encoding='utf-8') as r:
        while True:
            line=r.readline()
            if not line: break
            shop_list.append(line.strip())
    
    #json[intents]에 넣기
    star=json_data['intents'][0]
    star['tag']='market'
    star['patterns']=['매장추천해주세요','매장 무엇이 있나요?','매장 추천']
    star['responses']=shop_list
    
    #json에 수정
    with open('words.json','w',encoding='utf-8') as make_file:
        json.dump(json_data,make_file,ensure_ascii=False,indent='\t')
```

```python
import json
filename='brand_name.txt'
add_json(filename)
```
